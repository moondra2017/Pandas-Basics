`Boolean indexing`
---------------------

```python
df.loc[df.num_critic_for_reviews > 100, 'num_critic_for_reviews'].head()
```

```python
#& (and)  | (or), ~not
```

```python
df.loc[(df.num_critic_for_reviews > 100) & (df.director_facebook_likes> 200)].head()  #use brackets!
```

```python
df.loc[(df.num_critic_for_reviews == 100.0)].head() #exact match
```

```python
df.loc[~(df.num_critic_for_reviews > 100) & ~(df.director_facebook_likes> 100)].head()
```

```python
df.loc[(df.actor_2_name.str.startswith("V")) & ~(df.director_facebook_likes> 100)].head()
```

`Callable`
---------------------

```python
#Grab all rows where df.num_critic_for_reviews > 0 and all columns
```

```python
df.loc[lambda df: df.num_critic_for_reviews > 100, :]
```

```python
df.loc[: ,lambda df: ['num_critic_for_reviews']].head()
```
